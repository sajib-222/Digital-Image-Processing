```python
from google.colab import files
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
```

#### Read the image

```python
# Step 1: Provide the URL of the image
image_url = 'https://cdn.shopify.com/s/files/1/0274/9503/9079/files/20220211142754-margherita-9920_5a73220e-4a1a-4d33-b38f-26e98e3cd986.jpg?v=1723650067'  # Replace with the actual URL
```

```python
# Step 2: Download the image using urllib
with urllib.request.urlopen(image_url) as response:
    img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)  # Convert to NumPy array
```

```python
# Step 3: Decode the image (convert it into OpenCV format)
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)  # Decode as a color image (BGR format)
```

```python
# Step 4: Convert BGR to RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python

# Step 5: Display the image using Matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Remove axes for cleaner display
plt.title("Image from URL")
plt.show()
```

```python
height = img.shape[0]
width = img.shape[1]

print(height,width)
```

```python

img_crop=img_rgb[0:400,0:500]

plt.imshow(img_crop)
plt.axis('off')  # Remove axes for cleaner display
plt.title("Cropped Image")
plt.show()
```

## Select a 100x100 pixel region from the center of the image and display it separately.


```python
plt.imshow(img_rgb)
plt.axis('off')  # Remove axes for cleaner display
plt.title("Real Image")
plt.show()
```

```python
# Step 1: Compute center coordinates
center_x, center_y = width // 2, height // 2

# Step 2: Define crop boundaries (100x100 pixels)
start_x, end_x = center_x - 50, center_x + 50
start_y, end_y = center_y - 50, center_y + 50

# Step 3: Crop the region
center_crop = img_rgb[start_y:end_y, start_x:end_x]

# Step 4: Display the cropped region
plt.figure(figsize=(4, 4))
plt.imshow(center_crop)
plt.axis('off')  # Hide axes for clarity
plt.title("100x100 Center Crop")
plt.show()
```

# Flip the cropped image

cv2.flip(): This function flips the image either horizontally, vertically, or both, depending on the value passed as the second argument.

* First argument (img_crop): This is the image to be flipped.

* Second argument (-1): This is the flip code:

      - 0: Flip the image vertically (along the x-axis).
      - 1: Flip the image horizontally (along the y-axis).
      - -1: Flip the image both horizontally and vertically (i.e., a 180Â° rotation).


```python
img_flip=cv2.flip(img_crop,-1)  #flip horizontally and vertically

plt.imshow(img_flip)
plt.axis('off')  # Remove axes for cleaner display
plt.title("Cropped Image")
plt.show()
```

# Load an image, flip it upside down, rotate it 90Â° clockwise and counterclockwise, and display all versions

```python
# Load the image
pic = cv2.imread("/content/drive/MyDrive/DIP_Lab/Day_2/dog.jpg")

# Convert from BGR to RGB (Matplotlib expects RGB images)
pic = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)

# Flip and rotate the image
up = cv2.flip(pic, -1)  # Flip both vertically and horizontally
left = cv2.rotate(pic, cv2.ROTATE_90_CLOCKWISE)  # Rotate 90Â° clockwise
right = cv2.rotate(pic, cv2.ROTATE_90_COUNTERCLOCKWISE)  # Rotate 90Â° counterclockwise

# Function to display images using Matplotlib
def show_image(img, title):
    plt.imshow(img)
    plt.axis('off')  # Remove axes for a cleaner display
    plt.title(title)
    plt.show()

# Display images
show_image(pic, "Original Image")
show_image(up, "Flipped Image (Upside Down)")
show_image(left, "Rotated 90Â° Clockwise")
show_image(right, "Rotated 90Â° Counterclockwise")
```
