```python
import cv2
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
import pandas as pd
import seaborn as sns
from google.colab import files
```

### Show image from google drive

```python
image_path = '/content/drive/MyDrive/Digital Image Process/image/piza(2).jpg'
```

```python
img = cv2.imread(image_path)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.axis('off')
plt.title('RGB image')
plt.show()
```

### save image in google drive

```python
output_path = '/content/drive/MyDrive/Digital Image Process/image/output_image_of_piza.jpg'
cv2.imwrite(output_path, img_rgb)
print(f'Image saved at: {output_path}')
```

### Show Image from URL

```python
image_url = 'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRf6zoRR_FPG7f2knECoYTgOuETejMYPg71vg&s'
```

```python
# Step 2: Download the image using urllib
with urllib.request.urlopen(image_url) as response:
    img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)  # Convert to NumPy array
```

```python
image = cv2.imdecode(img_array, cv2.IMREAD_COLOR)  # Convert bytes to image
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.axis('off')
plt.title('RGB image')
plt.show()
```
