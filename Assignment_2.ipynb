```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = '/content/drive/MyDrive/Digital Image Process/image/desktop-wallpaper-4-anime-nature-anime-nature-pics.jpg'
```

```python
image = cv2.imread(image_path)
```

```python
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
def show_image(x,title):
  # Set figure size dynamically
  height, width, _ = x.shape
  fig_width = width / 100
  fig_height = height / 100
  plt.figure(figsize=(fig_width, fig_height))

  plt.imshow(x) #horizontal
  plt.title(title)
  plt.axis('off')
  plt.show()
```

## A. Resize an image to 256x256 pixels.



```python
show_image(image_RGB,'Original Image')
```

```python
img_resze = cv2.resize(image_RGB, (250, 250))
```

```python
show_image(img_resze,'Resized Image')
```

## B. Rotate by 180 degrees clockwise.


```python
img_rotate = cv2.rotate(image_RGB, cv2.ROTATE_180) # 180 degree clockwise
```

```python
show_image(img_rotate,'image rotate')
```

## C. Apply a vertical flip.

```python
img_flip_vertical = cv2.flip(image_RGB, 0)
```

```python
show_image(img_flip_vertical,'Vertical Flip')
```

## D. Display all transformations in a single subplot.

```python
plt.figure(figsize=(15, 15))

plt.subplot(1, 3, 1)
plt.imshow(img_resze)
plt.title('Resized Image to 256x256 pixels')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_rotate)
plt.title('Rotated Image 180 degrees clockwise')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img_flip_vertical)
plt.title('Vertical Flip')
plt.axis('off')

plt.tight_layout()
plt.show()
```
