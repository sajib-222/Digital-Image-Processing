```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
from google.colab import files
```

### **Task 1**: Extract a 100Ã—100 Region from the Center of the Image read from an URL.

```python
image_url = 'https://img.delicious.com.au/G-2mxbOh/w1200/del/2022/08/parmesan-crumbed-chicken-schnitzel-fried-eggs-and-apple-cabbage-slaw-173352-2.jpg'
```

```python
with urllib.request.urlopen(image_url) as request:
  image_array = np.asarray(bytearray(request.read()), dtype=np.uint8)
image = cv2.imdecode(image_array, cv2.IMREAD_COLOR)
```

```python
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
height, width, _ = image_RGB.shape

crop_width = 100
crop_height = 100

x1 = width // 2 - crop_width // 2
x2 = width // 2 + crop_width // 2
y1 = height // 2 - crop_height // 2
y2 = height // 2 + crop_height // 2

cropped_image = image_RGB[y1:y2, x1:x2]
plt.imshow(cropped_image)
plt.title('Cropped Image From Center')
plt.axis('off')
plt.show()
```

### **Task 2**:  Convert the image to HSV color space, visualize each components in grayscale representation.

```python
image_HSV = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(image_HSV)
plt.title('Image in HSV Color Space')
plt.axis('off')
plt.show()
```

```python
Hue, Saturation, Value = cv2.split(image_HSV)
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.imshow(Hue, cmap='gray')
plt.title('Hue Channel')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(Saturation, cmap='gray')
plt.title('Saturation Channel')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(Value, cmap='gray')
plt.title('Value Channel')
plt.axis('off')

plt.tight_layout()
plt.show()
```
