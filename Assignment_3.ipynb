```python
from google.colab import files
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

### Two 400x400 images with black background :

#### 1) First image contain a green rectangle at the center.

```python
image1 = np.ones((400, 400, 3), dtype=np.uint8)*0
cv2.rectangle(image1,(100,150),(300,250),(0,255,0),-1)
cv2_imshow(image1)
```

#### 2) Second image contain a magenta circle at the center.

```python
image2 = np.ones((400, 400, 3), dtype=np.uint8)*0
cv2.circle(image2, (200, 200), 100, (255, 0, 255), -1)
cv2_imshow(image2)
```

### AND

```python
bitwise_and = cv2.bitwise_and(image1, image2)
cv2_imshow(bitwise_and)
```

### OR

```python
bitwise_or = cv2.bitwise_or(image1, image2)
cv2_imshow(bitwise_or)
```

### XOR

```python
bitwise_xor = cv2.bitwise_xor(image1, image2)
cv2_imshow(bitwise_xor)
```

### NOT

```python
bitwise_not1 = cv2.bitwise_not(image1)
cv2_imshow(bitwise_not1)
```

### NOT

```python
bitwise_not2 = cv2.bitwise_not(image2) #negation
cv2_imshow(bitwise_not2)
```

### multiplication

```python
multiplication = cv2.multiply(image1, image2)
cv2_imshow(multiplication)
```

### Alpha blend

```python
alpha_blend = cv2.addWeighted(image1, 0.5, image2, 0.5, 0)
cv2_imshow(alpha_blend)
```

### subtraction

```python
subtraction = cv2.subtract(image1, image2)
cv2_imshow(subtraction)
```
